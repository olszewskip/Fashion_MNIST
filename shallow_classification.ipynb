```
import os
from joblib import dump, load
from random import randint
import numpy as np
import matplotlib.pyplot as plt

from keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split

from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```
path = '/content/gdrive/My Drive/Colab Notebooks/Fashion_MNIST'
```

```
(X_trainval, y_trainval), (_, _) = fashion_mnist.load_data()
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.1, random_state=42)

X20a_train = load(os.path.join(path,'rawX20a_train.joblib'))
X20a_val = load(os.path.join(path,'rawX20a_val.joblib'))
X20b_train = load(os.path.join(path,'rawX20b_train.joblib'))
X20b_val = load(os.path.join(path,'rawX20b_val.joblib'))
X20c_train_A = load(os.path.join(path,'rawX20c_train_A.joblib'))
X20c_train_B = load(os.path.join(path,'rawX20c_train_B.joblib'))
X20c_train_C = load(os.path.join(path,'rawX20c_train_C.joblib'))
X20c_train_D = load(os.path.join(path,'rawX20c_train_D.joblib'))
X20c_train_E = load(os.path.join(path,'rawX20c_train_E.joblib'))
X20c_train_F = load(os.path.join(path,'rawX20c_train_F.joblib'))
X20c_train = np.concatenate((X20c_train_A, X20c_train_B, X20c_train_C, X20c_train_D, X20c_train_E, X20c_train_F), axis=0).reshape(-1, 20)
X20c_val = load(os.path.join(path,'rawX20c_val.joblib')).reshape(-1, 20)
```

```
X20a_train.shape, X20a_val.shape, X20b_train.shape, X20b_val.shape, X20c_train.shape, X20c_val.shape, y_train.shape, y_val.shape
```
